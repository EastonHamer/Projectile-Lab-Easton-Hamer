```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
matplotlib.use('TkAgg')
%matplotlib inline



#Create arrays for your data

theta_array = [43.9-6,43.9-3,43.9,43.9+3,43.9+6]

y1 = [88.04]

y2 = [59.82]

y3 = [68.4]

y4 = [88.6]

y5 = [81.7]

ymean_array = [y5,y4,y1,y3,y2]
#Create an array for your y-axis uncertainties
yerr_array = (.1,.1,.1,.1,.1)
#Reassign variables
x = theta_array
y = ymean_array
dy = yerr_array

#size the plot
plt.figure(figsize=(15,10))

#create scatter plot
plt.scatter(x, y, color='blue', marker='o')

#create labels
plt.xlabel('$\\theta$ (degrees)')
plt.ylabel('$y_{mean}$ (m)')
plt.title('Height on wall vs Launcher Angle')

#fitting to a 2nd degree polynomial
c,b,a=np.polynomial.polynomial.polyfit(x,y,2,w=dy)

#Create fit line

xnew = np.linspace(min(x), max(x), 300)
fit = a*xnew**2 + b*xnew +c

plt.scatter(xnew, fit, color='red')
plt.show()

print ("C = ",c , "B = ",b, "A = ",a)
```
